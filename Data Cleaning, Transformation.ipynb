```python
import pandas as pd
import numpy as np

rohit1 = 'C://Users//H//Desktop//Spring18//-AdvancedDatabases//Project//labelled_data_Rohit.csv'
rohit2 = 'C://Users//H//Desktop//Spring18//-AdvancedDatabases//Project//labelled_data_Rohit1.csv'
abhi = 'C://Users//H//Desktop//Spring18//-AdvancedDatabases//Project//Project_Work//labelled_data.csv'
devosh = 'C://Users//H//Desktop//Spring18//-AdvancedDatabases//Project//Project_Work//labelled_tweets_devosh.csv'

rohit1 = pd.read_csv(rohit1, encoding='utf-8-sig')
rohit2 = pd.read_csv(rohit2, encoding='utf-8-sig')
devosh = pd.read_csv(devosh, encoding='utf-8-sig', sep=';;;')
abhi = pd.read_csv(abhi, encoding='utf-8-sig')
```

```python
# Cleaning

devosh1 = devosh[[' Tweet ', ' FavorOfGuns', ' State ']]
devosh1.columns = ['Tweet', 'For guns?', 'Location']
```

```python
# Combining datasets

final = [rohit1, rohit2, abhi, devosh1]
final = pd.concat(final)
final.dropna(inplace=True)
```

```python
final['For guns'] = np.where(final['For guns?'] == 1, 1, 0)
final['Against guns'] = np.where(final['For guns?'] == 0, 1, 0)
final['Makes no sense'] = np.where(final['For guns?'] == -2, 1, 0)
final
```

```python
final.loc[(final['For guns'] == 0) & (final['Against guns'] == 0) & (final['Makes no sense'] == 0)]
```

```python
final.to_csv('C://Users//H//Desktop//Spring18//-AdvancedDatabases//Project//Project_Work//final_dataset.csv', index=False,
         encoding='utf-8-sig')
```
