```python
import pandas as pd
import csv
import numpy as np
from textblob import TextBlob
```

```python
x=[]
with open('tweet_data.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        x.append(row['Tweets'])

Tweets = pd.DataFrame(x, columns = ['Tweets'])
```

```python
Tweets['sentiment'] = Tweets['Tweets'].apply(lambda tweet: TextBlob(tweet).sentiment)
```
